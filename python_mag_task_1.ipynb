Задача 1.
Напишите программу, которая принимает строку и возвращает количество гласных букв в ней. (Язык можете выбрать любой, с алфавитной письменностью)

```python
# ваш код
text = input()
vowels = ("a", "e", "i", "o", "u", "y")

counter = 0

for symb in text.lower():
  if symb in vowels:
    counter += 1

print(counter)
```

```python
# ваш код
text = input()
vowels = ("a", "e", "i", "o", "u", "y")

counter = 0

for letter in vowels:
  counter += text.lower().count(letter)

print(counter)
```

Задача 2. Напишите программу, которая удаляет повторяющиеся символы из строки, оставляя только первое вхождение каждого символа. Например, 'abracadabra' -> 'abrcd'

```python
# ваш код
text = input()

symb_set = set()
result = []

for symb in text:
  if not symb in symb_set:
    result.append(symb)
    symb_set.add(symb)

print("".join(result))
```

Задача 6. Как вы, надеюсь, уже знаете, метод count() находит количество подстрок в строке без учета пересечений. Например, в строке "аааа" подстрока "аа" будет найдена 2 раза. Напишите программу, которая считает, сколько раз подстрока встречается в строке с учетом пересечений. То есть, чтобы в строке "аааа" подстрока "аа" была найдена уже 3 раза.

```python
# ваш код
text = input()
sub_string_to_search = input()

counter = 0

for i in range(0, len(text) - len(sub_string_to_search) + 1):
  substr = text[i : i + len(sub_string_to_search)]
  if substr == sub_string_to_search:
    counter += 1

print(counter)
```

Задача 7. Напишите программу, которая принимает на вход список строк и возвращает новый список, где удалены дубликаты, но сохранен исходный порядок строк. Например, на список ["просто", "здравствуй", "просто", "как", "дела"] должен вернуться список ["просто", "здравствуй", "как", "дела"]

```python
# ваш код
text = input().split()

word_set = set()
result = []

for word in text:
  if not word in word_set:
    result.append(word)
    word_set.add(word)

print(result)
```

Задача 8. Напишите программу, которая принимает на вход строку с числами и находит в ней "счастливые" (те, у которых сумма цифр четная). Например, на строку "123 456 789 100" программа должна вернуть список [123, 789]

```python
# ваш код
numbers = input().split()

lucky_numbers_list = []

for number in numbers:
  digits = list(number)
  digits_sum = sum([int(digit) for digit in digits])
  if digits_sum % 2 == 0:
    lucky_numbers_list.append(number)

print(lucky_numbers_list)
```

Задача 10. Напишите программу, которая переведет строку из camelCase в kebab-case. Например, whatAYearThisWeekHadBeen будет переведена в what-a-year-this-week-had-been.

```python
# ваш код
text = input()

indexes = [0]

for i, symb in enumerate(text):
  if symb.isupper():
    indexes.append(i)

indexes.append(len(text))

# indexes = [i for i, symb in enumerate(text) if symb.isupper()]

new_text = ""

for index1, index2 in zip(indexes, indexes[1:]):
  new_text += text[index1 : index2].lower() + "-"

print(new_text[:-1])
```
