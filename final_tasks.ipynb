1. Возьмите текстовый файл с какой-нибудь книгой и постройте для нее частотный словарь в алфавитном порядке. Результат выведите в новый текстовый файл. В получившемся текстовом файле каждое встретившееся в книге слово должно выводиться в новой строке, через двоеточие - сколько раз это слово встретилось.

Пример:

а: 235

абажур: 2

аббат: 17

...

2. Напишите функцию, которая принимает на вход слово на русском языке с отмеченным ударением (значок выберите сами), а возвращает его же, но записанное по принципу "как слышится, так и пишется". Какие явления нужно отразить:

молоко -> малако

лесной -> лисной

мясной -> мисной

шести -> шысти

шипит -> шыпит

чаща -> чящи

цирк -> цырк

щука -> щюка


3. Напишите функцию, которая в качестве аргумента принимает путь к текстовому файлу, и выводит 5 самых длинных слов, встретившихся в тексте этого файла и то, сколько раз они встретились. Вывод нужно сделать в новый текстовый файл.

4. Напишите функцию, которая в качестве аргумента принимает путь к текстовому файлу, и выводит 5 самых длинных по числу слогов предложений, встретившихся в тексте файла. Число слогов считаем равным числу гласных. Считаем, что предложения в тексте могут заканчиваться только знаком ".". Вывод нужно сделать в новый текстовый файл.

5. Напишите программу, которая принимает на вход путь к текстовому файлу на русском языке, содержащему буквы "ё". Программа должна записать в новый файл этот же текст с "ё/Ё", замененными на "е/Е"

6. Напишите генератор случайных предложений. Заведите четыре текстовых файла: с существительными, глаголами, прилагательными и предлогами. Каждое слово в файлах пусть начинается с новой строки. Ваша функция должна генерировать случайное предложение из этих слов в форме: "существительное + глагол + предлог + прилагательное + существительное". Чтобы не мучиться с падежами, возьмите английский язык (но не забудьте тогда об артиклях и окончании глагола).

Пример вывода: "The cat sleeps on a red carpet"

7. Напишите функцию, которая принимает на вход путь к текстовому файлу и выводит, сколько в нем встретилось гласных, согласных, пробелов, цифр и знаков препинания.

8. Напишите программу, которая ищет все e-mail адреса в текстовом файле и выводит их в новый файл. 1 шаг: найдите все слова, включающие занк "@", 2 шаг: проверьте валидность адреса (это уже было в колабах), 3 шаг: запишите валидные адреса в новый файл.

9. Напишите программу, которая ковертирует строку в шифр Морзе для английского языка (достаточно букв). Пробелы обозначьте слэшем. https://ru.wikipedia.org/wiki/%D0%90%D0%B7%D0%B1%D1%83%D0%BA%D0%B0_%D0%9C%D0%BE%D1%80%D0%B7%D0%B5

Словарь морзянки должен читаться из текстового файла вида:

A	.-

B	-...

C -.-.

Пример вывода: "... --- ... / .--. -.-- - .... --- -."

10. Напишите программу, которая принимает на вход путь к текстовому файлу с текстом на английском языке и выводит в новый файл все пары слов, встретившихся в тексте, которые будут выглядеть одинаково, если перевернуть одно из них кверх ногами. Пример: "pod" ↔ "boq". Используйте симметричные пары букв p ↔ q, b ↔ d, n ↔ u, o ↔ o, s ↔ s, z ↔ z

12. Напишите функцию поиска цитат в тексте. На вход она должна принимать имя файла, а в новый файл выводить все цитаты в порядке следования в тексте с нумерацией. Цитатой считается часть текста, заключенная в двойные кавычки.

Пример текста: Он сказал: "Привет", а потом "Пока".
