Задача 1. Напишите программу, которая проверяет, является ли слово изограммой (все буквы уникальны, без учета регистра).

```python
word = input()

if len(word) == len(set(word)):
  print(f"Слово {word} является изограммой")
else:
  print(f"Слово {word} не является изограммой")
```

Задача 3. Найдите символ, который встречается в строке чаще всего (если их несколько, верните любой).

```python
sentence = list(input())

char_counter = {}
for character in sentence:
  char_counter[character] = sentence.count(character)

print(char_counter)
print(char_counter.keys(), max(char_counter.keys()))

max_count = max(char_counter.values())

max_counted_chars = [lady for lady, gaga in char_counter.items() if gaga == max_count]

print(max_counted_chars, max_count)
```

Задача 4. Объедините два словаря, суммируя значения для одинаковых ключей. Например, для словарей

dict1 = {'a': 1, 'b': 2}

dict2 = {'b': 3, 'c': 4}

результатом будет {'a': 1, 'b': 5, 'c': 4}

```python
dict1 = {'a': 1, 'b': 2}

dict2 = {'b': 3, 'c': 4}

# dict1.update(dict2)

dict3 = {}

for key, value in dict2.items():
  if key in dict1.keys():
    dict3[key] = dict2[key] + dict1[key]

print(dict1, dict2, dict3)

dict1.update(dict2)
print(dict1, dict2, dict3)

dict1.update(dict3)
print(dict1, dict2, dict3)

print(dict1)
```

```python
dict1 = {'a': 1, 'b': 2}

dict2 = {'b': 3, 'c': 4}

# dict1.update(dict2)

dict3 = {}

for key, value in dict2.items():
  if key in dict1.keys():
    dict3[key] = dict2[key] + dict1[key]
  else:
    dict3[key] = value

for key, value in dict1.items():
  if  key not in dict3.keys():
    dict3[key] = value

dict3
```

Задача 5. Напишите программу, которая принимает две строки одинаковой длины и возвращает количество позиций, в которых символы различаются (расстояние Хэмминга).

```python
str_1 = list(input())
str_2 = list(input())

counter = 0

for i, char in enumerate(str_1):
  print(i, char, str_2[i], char != str_2[i])
  if char != str_2[i]:
    counter += 1

print(counter)
```

```python
str_1 = list(input())
str_2 = list(input())

counter = 0

for i in range(len(str_1)):
  print(i, str_1[i], str_2[i], str_1[i] != str_2[i])
  if str_1[i] != str_2[i]:
    counter += 1

print(counter)
```
