Задание 2. Ваша программа гененрирует серийные номера роботов из 6 цифр. Напишите функцию, определяющую, является ли номер робота счастливым (сумма первых трех цифр равна сумме следующих трех цифр).

1. Напишите функцию, определяющую, является ли номер счастливым
2. Сгенерируйте шестизначный номер робота с помощью random.randint модуля random
3. Выведите номер робота, проверьте является ли он счастливым, и тоже выведите это на экран.

```python
def sum_digits(number):
  first_sum = 0
  for digit in number:
    first_sum += int(digit)
  return first_sum

def is_number_lucky(number):
  number = str(number)
  first_digits = number[:3]
  last_digits = number[3:]
  # first_sum = int(first_digits[0]) + int(first_digits[1]) + int(first_digits[2])
  # second_sum = int(last_digits[0]) + int(last_digits[1]) + int(last_digits[2])

  first_sum = sum_digits(first_digits)
  second_sum = sum_digits(last_digits)

  return first_sum == second_sum
```

```python
robot_number = random.randint(100000, 999999)
print("Robot number: ", robot_number)
print("Is lucky: ", is_number_lucky(robot_number))
```

Задание 3. По сюжету игры, которую вы решили написать, на пути вашего героя случается мост, по которому не дает пройти его хранитель. Хранитель хочет, чтобы герой ответил на три его вопроса.

1. Задайте словарик вида {вопрос: ответ} с тремя вопросами.
2. В цикле задайте вопросы по очереди, запрашивая ответы через input().
3. Если герой ошибся, выведите сообщение: "Вы упали в пропасть!", если ответил на все три вопроса: "Вы прошли мост!"

```python
questions_dict = {
    "Как дойти до неизведанной деревни ": "Нужно следовать за вороном",
    "Какова скорость ненагруженной ласточки ": "Европейской или африканской?",
    "Сколько будет 3 + 3 ": "6"
}

is_right = False

for key, value in questions_dict.items():
  answer = input(key)
  is_right = answer == value
  if not is_right:
    print("Вы упали в пропасть!")
    break

if is_right:
  print("Вы прошли мост!")
```

Задание 5.

Охота за сокровищами. Напишите мини-игру со следующим алгоритмом:

1. Программа загадывает координаты x, y сокровища на карте 5х5. Сделать это можно с помощью функции random.randint(1, 5). Не забудьте импортировать модуль random.
2. Игрок в цикле вводит свои координаты x, y
3. Если игрок угадал координаты, программа выдает фразу "Ты нашел сокровища!"
4. Если сумма расстояний между загаданными координатами и координатами игрока меньше двух, программа пишет "Горячо!"
5. Иначе - пишет "Холодно..."
6. Цикл продолжается, пока игрок не угадает координаты сокровища.

```python
X = random.randint(1, 5)
Y = random.randint(1, 5)

while True:
  x_user = int(input("Введите координату по горизонтали: "))
  y_user = int(input("Введите координату по вертикали: "))

  if X == x_user and Y == y_user:
    print(f"Ты нашел сокровища в клетке ({X},{Y})!")
    break
  elif abs(X - x_user) + abs(Y - y_user) <= 4:
    print(f"Горячо!")
  else:
    print("Холодно...")
```
