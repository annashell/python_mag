Если использовать *args, аргументов может быть сколько угодно:

```python
def make_sum_2(*numbers):
  return sum(numbers)
```

```python
make_sum_2(1, 2, 3)
```

```python
make_sum_2(3, 7, 1, 4, 7)
```

```python
make_sum_2()
```

Если после *args задается еще какой-то аргумент, его можно будет ввести только через ключевое слово:

```python
def make_sum_3(*numbers, coeff):
  return coeff * sum(numbers)
```

```python
make_sum_3(3, 4, 1, coeff=2)
```

```python
make_sum_3(3, 4, 1)
```

Можно задать дефолтное значение аргумента, которое будет использовано, если при вызове функции значение этого аргумента не указано. Если дефолтное значение для аргумента задано, он становится не обязательным.

```python
def make_sum_4(*numbers, coeff=1):
  return coeff * sum(numbers)
```

```python
make_sum_4(3, 4, 1)
```

```python
make_sum_4(3, 4, 1, coeff=2)
```

Символ * в Python означает распаковку, например:

```python
a = [1, 2, 3]
b = [*a, 4, 5, 6]
b
```

Если *args в качестве аргумента задает позиционные аргументы в форме кортежа, то * *kwargs - именованные (keyword arguments) в форме словаря:

```python
def func(*args, **kwargs):
    print(args)  # это кортеж
    print(kwargs)  # а это словарь!
```

```python
func(1, 2, 3)
```

```python
func(1, 2, a=1, b=3, c=45)
```

```python
def print_group_scores(group, **students):
   print(f"Группа: {group}\n")
   for student, score in students.items():
      print(f"{student}: {score}")
```

```python
print_group_scores("123", Vasya=4.5, Petya={4: 5}, Danya=3, Vanya="78")
```

```python
c = print_group_scores("123", Vasya=4.5, Petya={4: 5}, Danya=3, Vanya="78")
print("c", c)
```

Задача. Напишите функцию, которая на вход принимает список слов и последовательность символов, а возвращает те слова, которые содержат эту последовательность символов.

```python
def filter_word_list_by_sequence(*words, sequence="аб"):
  print(words, sequence)
  result_list = []

  for word in words:
    if sequence in word:
      result_list.append(word)

  # result_list = [word for word in words if sequence in word]

  return result_list
```

```python
filter_word_list_by_sequence("абрикос", "кокос", "лаборатория", "персик")
```

```python
filter_word_list_by_sequence("абрикос", "кокос", "лаборатория", "персик", sequence="ко")
```
